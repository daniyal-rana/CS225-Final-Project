We used the above website to calculate [Hollinger's PER](https://en.wikipedia.org/wiki/Player_efficiency_rating#Calculation)

Because we are estimating Hollinger's PER using the above estimate, we want the difference between our calculated PER values vs. online PER values to be within a margin of error of 2 points.

```python
import pandas as pd

df = pd.read_csv('NBA_STATS_WITH_PER.csv')
df
```

```python
#check players PER properly calculated

import random

seed_val = 225 #used to select random player

for i in range(5):
    random.seed(seed_val)
    index = random.randint(0, len(df))
    #print(index)
    player = df.loc[index]
    print( i+1, " Name ",player['Player'], "Team ",player['Tm'], " Year ",player['Year']," PER ", player['PER'])
    seed_val+=1
```
